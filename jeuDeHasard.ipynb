```julia
function jouerNParties(nbSemaines) #nbParties est le nombre de parties qu'on veut jouer (simuler). C'est un paramètre, on ne
    #connait pas sa valeur au moment de la définition de la fonction. Mais quand on va appeler la fonction jouerNParties()
    #on doit donner une valeur à nbParties (par exemple, si on appelle jouerNParties(100), le programme va s'exécuter
    #en remplaçant nbParties par la valeur 100) 
    sommeNbVentes = 0 #Va contenir la somme, sur toutes les parties, du nombre de tirages nécessaires pour finir une partie
    println("Semaine | nombre de vente")
    esperence=7
    for i = 1:nbSemaines #Chaque i est une partie
        
        random=rand(Float64)
        nbVentes=random*esperence
        
        @printf("%6d %11d\n", i, nbVentes)        
        #@printf est utilisée comme la fonction printf du langage C. Ici, par exemple, dans "%6d %11d %13d\n", le premier %6d 
        #sera remplacé par la première variable qui apparait après les guillemets et après la virgule dans l'appel de printf. 
        #Donc le résultat est de remplacer %6d par la valeur de la variable i. Le nombre 6 est en fait le nombre d'espace sur
        #lequel on écrit la valeur de la variable i. La lettre d veut dire qu'on affiche un nombre entier. Donc la commande
        #@printf("%6d %11d %13d\n", i, nbTirages, (argentRecuFin - prixTotal)) va afficher la valeur de la 
        #variable i sur 6 espaces, puis une espace, puis la valeur de la variable nbTirages sur 11 espaces (colonnes), puis
        #une espace, puis la valeur du résultat (argentRecuFin - prixTotal) sur 13 colonnes puis imposera un retour à la 
        #ligne. Le retour à la ligne est le résultat de l'écriture \n
        
        sommeNbVentes += nbVentes              #Mettre à jour la valeur de la variable sommeNbTirages
    end  #Correspond à la boucle for i = 1:nbParties
    
    println("Nombre de vente (en moyenne) : ", sommeNbVentes/nbSemaines)
    
end #Correspond à la fin de la déclaration de la fonction
```

```julia
#Simuler 14 parties (avec quelques statistiques)
jouerNParties(30)
```

```julia
#Simuler 1000 parties (avec quelques statistiques)
jouerNParties(300)
```
